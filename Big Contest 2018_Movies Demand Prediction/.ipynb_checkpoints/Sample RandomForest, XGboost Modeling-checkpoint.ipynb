```python
xgb_params = {
    'objective': ['reg:logistic', 'reg:gamma'],
    'learning_rate': [0.001, 0.01, 0.1, 0.2],
    'max_depth': [4,6,8],
    'gamma' :[0.01, 0.1, 1]
}
```
