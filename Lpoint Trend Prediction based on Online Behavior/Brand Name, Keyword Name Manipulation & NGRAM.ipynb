```python
def diff_ngram2(sa,sb):
    res1=[]
    res2=[]
    
    if(pd.isnull(sa)==True | pd.isnull(sb)==True):
        return 0
    
    else:
        if(len(sa)<3 | len(sb)<3):
            num=min(len(sa),len(sb))
        else:
            num=3

        alen=len(sa)-num+1
        blen=len(sb)-num+1
        for i in range(alen):
            aa=sa[i:i+num]
            res1.append(aa)
        for i in range(blen):
            bb=sb[i:i+num]
            res2.append(bb)
        cnt=0
        for i in res1:
            for j in res2:
                if i==j:
                    cnt+=1
        return cnt/len(res1)
```

```python
def diff_ngram2(sa,sb,num):
    res1=[]
    res2=[]
    alen=len(sa)-num+1
    blen=len(sb)-num+1
    for i in range(alen):
        aa=sa[i:i+num]
        res1.append(aa)
    for i in range(blen):
        bb=sb[i:i+num]
        res2.append(bb)
    r=[]
    cnt=0
    for i in res1:
        for j in res2:
            if i==j:
                cnt+=1
                r.append(i)
    return cnt/len(res1),r
```
