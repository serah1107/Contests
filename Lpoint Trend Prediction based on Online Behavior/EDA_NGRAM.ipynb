```python
a=total["PD_BRA_NM"]+total["PD_NM"]
b=total["KWD_NM"]
```

```python
def diff_ngram2(sa,sb,num):
    res1=[]
    res2=[]
    alen=len(sa)-num+1
    blen=len(sb)-num+1
    for i in range(alen):
        aa=sa[i:i+num]
        res1.append(aa)
    for i in range(blen):
        bb=sb[i:i+num]
        res2.append(bb)
    r=[]
    cnt=0
    for i in res1:
        for j in res2:
            if i==j:
                cnt+=1
                r.append(i)
    return cnt/len(res1),r
```

```python
len(str(b[1]))
```

```python
len(b)
```

```python
ngram=[]
for i in range(73341102):
    if len(str(b[i]))== 2:
        k=diff_ngram2(str(a[i]),str(b[i]),2)
        
    elif len(str(b[i]))== 1:
        k=diff_ngram2(str(a[i]),str(b[i]),1)
        
    elif len(str(b[i]))== 0:
        k="no search" 
    else:
        k=diff_ngram2(str(a[i]),str(b[i]),3)
    ngram.append(k)
```
